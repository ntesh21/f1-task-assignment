```python
import cv2
import face_recognition
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load the document image

```python
def load_document(path):
    doc_img = cv2.imread(path)
    return doc_img
    
```

## Convert to Grayscale

```python
def convert_grayscale(img):
    img = cv2.cvtColor(img,
                   cv2.COLOR_BGR2GRAY)
    return img
```

## Detect the face from the Document

```python
def detect_faces(img):
    #Load haarcascade model
    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    #Detect face with haarcascade model
    faces = faceCascade.detectMultiScale(
        img,
        scaleFactor=1.3,
        minNeighbors=3,
        minSize=(30, 30)
    ) 
    if len(faces) > 0:
        print("Found {0} Faces!".format(len(faces))) 
    else:
        print('No face found')
    return faces 


    
```

## Extract the detected face

```python
def extract_face(img, faces):
    total_faces = []
    if len(faces) > 0:
        for (x, y, w, h) in faces:
            cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
            roi_color = img[y:y + h, x:x + w] 
            total_faces.append(roi_color) 
            cv2.imwrite('taskB-extracted-images/'+str(w)+'_extracted_face.jpg', roi_color)
        cv2.imwrite('taskB-extracted-images/face_detected.jpg', img)
        plt.figure(figsize=(11,6))
        rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.subplot(121),plt.imshow(rgb_img, cmap = 'gray')
        plt.title('Detected Face'), plt.xticks([]), plt.yticks([])
        plt.show()  # display it
        return total_faces
    else:
        return None
        
    
```

## Main method to get the face image from the input document

```python
def main(path):
    img = load_document(path)
    gray = convert_grayscale(img)
    faces = detect_faces(gray)
    face_extracted = extract_face(img, faces)
    return face_extracted
```

```python
extracted_face = main('task-B.jpeg')
```

## Method to compare the live image with extracted image

```python
def compare_faces(img1, img2):
    # Load live image to compare against
    live_image = img1

    # Get the face encodings for the known images
    live_image_encoding = face_recognition.face_encodings(live_image)[0]

    known_encodings = [
        live_image_encoding,
    ]

    # Load a extracted image from the document
    image_to_test = img2
    image_to_test_encoding = face_recognition.face_encodings(image_to_test)[0]

    #Find is the images are of same person or not
    results = face_recognition.compare_faces([known_encodings[0]], unknown_encoding)
    print("Are the both images of same person? \n {}".format(results[0]))
    
    # See how far apart the test image is from the known faces
    face_distances = face_recognition.face_distance(known_encodings, image_to_test_encoding)

    for i, face_distance in enumerate(face_distances):
        print("The extracted image has a distance of {:.2} from live image.".format(face_distance))
        print("- With a normal cutoff of 0.6, would the test image match the known image? \n {}".format(face_distance < 0.6))
        print("- With a very strict cutoff of 0.5, would the test image match the known image? \n {}".format(face_distance < 0.5))
        print()
    return face_distances[0]
```
